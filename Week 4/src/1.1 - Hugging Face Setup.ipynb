# Hugging Face Token Setup Tutorial

This notebook will guide you through the process of setting up your Hugging Face access token for Week 4 exercises.

## Step 1: Visit Hugging Face

Go to [https://huggingface.co](https://huggingface.co) and sign up for a free account if you don't have one yet.

![Hugging Face Homepage](../assets/hugging_face_setup_images/hugging_face_1.png)

## Step 2: Log In to Your Account

Click **Log In** in the top-right corner and enter your username/email and password.

![Hugging Face Login](../assets/hugging_face_setup_images/hugging_face_2.png)

## Step 3: Open Your Profile Menu

Once logged in, click on your profile avatar in the top-right corner to open the dropdown menu, then click **Access Tokens**.

![Profile Menu — Access Tokens](../assets/hugging_face_setup_images/hugging_face_3.png)

## Step 4: Create a New Token

On the **Access Tokens** page you can see all your existing tokens. Click **+ Create new token** to generate one for this project.

![Access Tokens Page](../assets/hugging_face_setup_images/hugging_face_4.png)

## Step 5: Choose the Token Type

Select the token type. For most Week 4 exercises a **Read** token is sufficient — it grants read-only access to models and datasets. Select **Read** from the three options (Fine-grained / Read / Write).

![Token Type Selection](../assets/hugging_face_setup_images/hugging_face_5.png)

## Step 6: Name Your Token and Create It

Give your token a descriptive name (e.g., `WMG_AI`) and click **Create token**.

![Name and Create Token](../assets/hugging_face_setup_images/hugging_face_6.png)

## Step 7: Copy Your Token

**Important**: Copy your token immediately — it will only be shown **once**! Click the **Copy** button and store it somewhere safe before closing this dialog.

![Save Your Access Token](../assets/hugging_face_setup_images/hugging_face_7.png)

## Step 8: Create the .env File

Inside the `Week 4` folder, create a file named `.env`


![Environment File](../assets/hugging_face_setup_images/hugging_face_8.png)

## Step 9: Add to Environment File

Add your token to the `Week 4/.env` file:

```
HF_TOKEN=your_actual_token_here
```

![Environment File](../assets/hugging_face_setup_images/hugging_face_9.png)

## Verify Setup

Run the cell below to confirm your token is properly configured.

````python
import os
from dotenv import load_dotenv
from huggingface_hub import InferenceClient
from pathlib import Path
from munch import Munch

load_dotenv("../.env")

HF_TOKEN = os.getenv("HF_TOKEN")

# Load config
config_path = (Path.cwd() / "../config/config.yaml").resolve()
with open(config_path, "r") as f:
    config = Munch.fromYAML(f)

if not HF_TOKEN:
    print("❌ HF_TOKEN not found. Please add it to your .env file.")
else:
    print("✅ HF_TOKEN found in environment")

    try:
        client = InferenceClient(
            model=config.hugging_face.model,
            token=HF_TOKEN
        )

        response = client.chat_completion(
            messages=[{"role": "user", "content": "Reply with one short sentence: what is 2+2?"}]
        )

        answer = response.choices[0].message.content
        print("✅ LLM call successful!")
        print(f"Response: {answer}")

    except Exception as e:
        print(f"❌ LLM call failed: {e}")
````
